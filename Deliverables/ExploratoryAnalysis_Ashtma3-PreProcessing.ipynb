```python
data = pd.read_csv('../Datasets/Disease cases/asthma-3/asthma-ed-visit-rates-by-county.csv', encoding="ISO-8859-1",low_memory =False)
data.head()
```

```python
data['YEAR'] = data['YEAR'].map(lambda x: int(x))
data.head()
```

```python
#county population file
county_population = pd.read_csv('../Datasets/Disease cases/asthma-4/county_population.csv',low_memory=False)

#remove comma from Population column
county_population['Population'] = county_population['Population'].str.replace(',','').astype(int)
county_population.head()
```

```python
merged = pd.merge(data,county_population,how='inner', left_on=['YEAR', 'COUNTY'], right_on=['Year','County'])
merged.head()
```

```python

merged['Patient_ratio'] = merged['NUMBER OF ED VISITS']/merged['Population']
merged.tail()
```

```python
corr = merged.corr()
g = sns.heatmap(corr, vmax=.9, center=0,
               square=True, linewidths=.9, cbar_kws={"shrink": .8}, annot=True, fmt='.2f', cmap='coolwarm')
bottom, top = g.get_ylim()
g.set_ylim(bottom + 0.5, top - 0.5)
sns.despine()
g.figure.set_size_inches(16, 10)
```

```python
filter_county = ['Napa', 'Marin', 'Sonoma', 'Solano', 'Alameda', 'San Francisco',
       'Contra Costa', 'Santa Clara']
 
v = merged[( merged['COUNTY'].isin(filter_county)) & (merged['NUMBER OF ED VISITS'].notnull())]
v
```
