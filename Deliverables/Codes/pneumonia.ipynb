```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
pneumonia = pd.read_csv('../Datasets/Disease cases/pneumonia/pneumonia-clean.csv', low_memory =False)
pneumonia.tail()
```

```python
county_population = pd.read_csv('../Datasets/Disease cases/asthma-4/county_population.csv',low_memory=False)
county_population['Population'] = county_population['Population'].str.replace(',','').astype(int)
county_population.tail()
```

```python
merged = pd.merge(pneumonia,county_population,how='inner', left_on=['YEAR', 'COUNTY'], right_on=['Year','County'])
merged.tail()
```

```python
merged.dtypes
```

```python
merged['Patient_ratio_deaths'] = merged['# of Deaths'] / merged['Population']
merged['Patient_ratio_cases'] = merged['# of Cases'] / merged['Population']
merged.head()
```

```python
filter_county = ['Napa', 'Marin', 'Sonoma', 'Solano', 'Alameda', 'San Francisco','Contra Costa', 'Santa Clara']
filtered_pneumonia = merged[( merged['COUNTY'].isin(filter_county))]
filtered_pneumonia
```
