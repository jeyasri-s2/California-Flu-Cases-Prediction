```python
# Manmeet Singh
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline

def main():
    path = ("./Datasets/Disease_cases/asthma-3/asthma-ed-visit-rates-by-county.csv")
    data = pd.read_csv(path, encoding = "ISO-8859-1")
    visits_by_year=data.groupby('YEAR')['NUMBER OF ED VISITS'].sum()
    visits_by_year.plot(kind='bar')
    
if __name__=="__main__":
    main()
```

```python
data = pd.read_csv(path, encoding = "ISO-8859-1")
vists_by_county=data.groupby('COUNTY')['NUMBER OF ED VISITS'].sum()
vists_by_county.plot(kind='bar', figsize=[16,12])
```
