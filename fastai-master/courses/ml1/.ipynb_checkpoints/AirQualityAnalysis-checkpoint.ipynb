```python
data = pd.read_csv('data/AQData_monthly_2014_2019.csv', low_memory= False)
```

```python
data.head()
```

```python

data['ZoneCategory'] = data.Zone.astype("category").cat.codes
data['ZoneCategory']

```

```python
data['StationCategory'] = data.Station.astype("category").cat.codes
data['StationCategory']
```

```python
flds = [ 'ZoneCategory', 'MonthNo','Value']
enc_summ = data[flds].groupby('MonthNo', as_index=False).count()
enc_summ
```

```python
plt.scatter(data['ZoneCategory'], data['Value'])
plt.show()
```

```python
fields = ['Zone','ZoneCategory']
data[fields].groupby('Zone', as_index=False).count()
```

```python
station_fields = ['Station','StationCategory']
data[station_fields].groupby('Station', as_index=False).count()
```

```python
zstation_fields = ['Zone','Station','MonthNo']
data[zstation_fields].groupby(['Zone','Station'], as_index=False).count()
```
